# Polynomial Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("polynomial_linear_regression.csv", sep = ";")
```

```python
df.head()
```

```python
x = df.araba_fiyat.values.reshape(-1, 1)
y = df.araba_max_hizi.values.reshape(-1, 1)
```

```python
plt.scatter(x, y)
plt.xlabel("araba_fiyat")
plt.ylabel("araba_max_hizi")
plt.show()
```

```python
# linear regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x, y)
```

```python
# predict
y_head = lr.predict(x)

plt.scatter(x, y)
plt.ylabel("araba_max_hizi")
plt.xlabel("araba_fiyat")

plt.plot(x, y_head, color = "red", label ="linear")
#plt.show()

print("10M tl lik araba hizi:",lr.predict([[10000]]))
```

```python
# polynomial_linear_regression -> y = b0 + b1*x + b2*x^2 + ... + bn*x^n

from sklearn.preprocessing import PolynomialFeatures

polynomial_regression = PolynomialFeatures(degree=2)
x_polynomial = polynomial_regression.fit_transform(x)
```

```python
# fit
linear_regression2 = LinearRegression()
linear_regression2.fit(x_polynomial,y)
```

```python
y_head2 = linear_regression2.predict(x_polynomial)

plt.scatter(x, y)
plt.xlabel("araba_fiyat")
plt.ylabel("araba_max_hizi")

plt.plot(x, y_head, color = "red", label ="linear")
plt.plot(x, y_head2, color = "green", label = "poly")
plt.legend()
plt.show()
```
