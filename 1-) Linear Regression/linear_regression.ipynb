```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("linear_regression_dataset.csv", sep = ";")
```

```python
df.head()
```

```python
plt.scatter(df.deneyim, df.maas)
plt.xlabel("deneyim")
plt.ylabel("maas")
plt.show()
```

Yukaridaki gorselde x ekseni deneyim, y ekseni ise maasi gostermekte. Deneyim arttikca maas da artmakta. Bu gorselde noktalara baktigimizda zihnimizde bir cizgi canlanmakta. Bunu modele cevirebilmemiz icin mavi noktalara line fit etmeliyiz ve bu sayede istedigimiz degeri tahmin edebiliriz. Buna "line fit" veya "linear regression" denir. 

Bu ifadenin matematiksel karsiligi;

y = b0 + b1 * x 'dir.

>y: deneyim, 

>x: maas, 

>b0: bias degeri, 

>b1: egim


maas = b0 + b1 * deneyim



Yukaridaki gorselde yaptigimiz line fit isleminde bu islemi ne kadar iyi yaptik, bu islemde ne kadar hata var onu bulmak icin asagidaki islemleri takip edecegiz:

residual = y - y_head

>y_head: tahmin edilen deger

MSE = sum(residual ** 2) / n
>mse: mean square error

>n: sample sayisi

amacimiz ortalama hatayi en dusuk tutumaktir.

matematiksel ifadesi;


```python
# linear regression

# sklearn library
from sklearn.linear_model import LinearRegression

# linear regression model
linear_reg = LinearRegression()

x = df.deneyim.values.reshape(-1,1)
y = df.maas.values.reshape(-1,1)

linear_reg.fit(x, y)
```

```python
# prediction

b0 = linear_reg.predict([[0]])
print("b0: ", b0)

b0_ = linear_reg.intercept_  # y eksenini kestigi nokta, intercept
print("b0_: ", b0_)
```

```python
b1 = linear_reg.coef_ # egim, slope
print("b1: ", b1)
```
