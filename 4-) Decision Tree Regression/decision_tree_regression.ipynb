# Decision Tree Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("decision_tree_regression_dataset.csv", sep = ";", header = None)
```

```python
df. head()
```

```python
x = df.iloc[:,0].values.reshape(-1, 1)
y = df.iloc[:,1].values.reshape(-1, 1)
```

```python
# decision tree regression
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(x, y)
```

```python
tree_reg.predict([[5.5]])
#x_ = np.arange(min(x), max(x), 0.01).reshape(-1, 1)
y_head = tree_reg.predict(x)
```

```python
# visualize
plt.scatter(x, y, color = "red")
plt.plot(x, y_head, color = "green")
plt.xlabel("tribun level")
plt.ylabel("ucret")
plt.show()
```

Yukarida tribun level azaldikca ucrette azalmakta ornegin 1' den 2' ye gecis adimlarinda, halbuki tribun level gecislerde ucret sabit olmasi gerekir. Bunun icin asagidaki adimlari uyguluyoruz.

```python
x_ = np.arange(min(x), max(x), 0.01).reshape(-1, 1)
y_head = tree_reg.predict(x_)
```

```python
# visualize
plt.scatter(x, y, color = "red")
plt.plot(x_, y_head, color = "green")
plt.xlabel("tribun level")
plt.ylabel("ucret")
plt.show()
```

Gorselde goruldugu gibi tribun levellerde ucret sabit, gecislerde ucret degismekte.
